```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
from pprint import pprint as pp
from pyarrow import parquet as pq
import pywt
from scipy import signal as dsp
import seaborn as sns
from sklearn.externals import joblib

sns.set()
```

```python
DATA_DIR = "../data/"
NUM_LINES = 500

test_start = 8712  # TODO: What is this?
```

```python
train = pq.read_pandas(
    Path(DATA_DIR, 'train.parquet'),
    columns=[str(i) for i in range(NUM_LINES)]
).to_pandas().values.T
test = pq.read_pandas(
    Path(DATA_DIR, 'test.parquet'),
    columns=[str(i) for i in range(test_start, test_start + NUM_LINES)]
).to_pandas().values.T
meta_train = pd.read_csv(Path(DATA_DIR, 'metadata_train.csv'))
meta_test = pd.read_csv(Path(DATA_DIR, 'metadata_test.csv'))
```

```python
idx = np.random.randint((train.shape[0] / 3) + 1,)

fig, ax = plt.subplots(3, 1, sharex='col')
measurements = list()
for count in range(3):
    plt.subplot(((count + 1) * 1) + 310)
    plt.plot(train[idx + count])
    measurements.append(meta_train.iloc[idx + count].id_measurement)
fig.subplots_adjust(hspace=0.4, wspace=0.4)
plt.suptitle("Training Examples: {} {}".format(meta_train.iloc[idx + count].target,
                                               measurements))
```

```python
sig = train[idx + 1]
target = meta_train.iloc[idx + 1].target
print("PD: {}".format(meta_train.iloc[idx + 1].target))

cA, cD = pywt.dwt(sig, 'db3')

# %matplotlib notebook
%matplotlib inline
plt.plot(cD)
plt.plot(cA)
plt.show()
```

```python
sig_mean = np.absolute(cD).mean()
args = { 'height': 2.5, 'threshold': 2.5, 'distance': 1000, 'prominence': 5, 'width': 1, 'wlen': 300 }
args = {
    'height': sig_mean + 3.5,
    'prominence': 2.5,
    'distance': 700
}
peaks, pinfo = dsp.find_peaks(cD, **args)
valleys, vinfo = dsp.find_peaks(-cD, **args)

%matplotlib inline
plt.figure(figsize=(22, 7))
plt.plot(cD, ls='-')
plt.plot(cD, 'o', color='r', markevery=list(peaks))
plt.plot(cD, 'o', color='r', markevery=list(valleys))
title = "SigMean: {:1.2}, Peaks: {}, Valleys: {}, Target: {}".format(sig_mean, len(peaks), len(valleys), target)
plt.title(title, fontsize="28")
plt.show()
```

```python
plt.hist(abs(cD), bins=100, range=(0, 10))
```

```python
sub_peaks = peaks[0:4]
# proms = dsp.peak_prominences(sig, sub_peaks, wlen=300)
proms = dsp.peak_prominences(sig, sub_peaks)
print(sub_peaks)
if 'prominences' in pinfo:
    print(pinfo.get('prominences')[0:4])
print(proms)
```
